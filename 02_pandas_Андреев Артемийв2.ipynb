3.1  Добавьте в таблицу `recipes` столбец `description_length`, в котором хранится длина описания рецепта из столбца `description`.

```python
recipes['description_length'] = recipes['description'].apply(lambda x: len(str(x)))

print(recipes.head())
```

Задание 3.2

```python
recipes['name'] = recipes['name'].apply(lambda x: x.title()) 

print(recipes.head())
```

Задание 3.3

```python
recipes['name_word_count'] = recipes['name'].apply(lambda x: len(list(filter(None, x.split(" "))))) 

print(recipes.head())
```

Задание 4.1

```python
recipes_count = recipes.groupby('contributor_id').size().reset_index(name='recipe_count')

max_recipe = recipes_count[recipes_count['recipe_count'] == recipes_count['recipe_count'].max()]

print(max_recipe)
```

Задание 4.2

```python
average_rating = reviews.groupby('recipe_id')['rating'].mean().reset_index()

valid_reviews = reviews[(reviews['rating'] != 0) & (reviews['review'].notnull())]

recipes_without_reviews = recipes[~recipes['id'].isin(valid_reviews['recipe_id'].unique())]

print(average_rating)

print(recipes_without_reviews)
```

Задание 5.1

```python
merged = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='inner') 

merged = merged[['id', 'name', 'user_id', 'rating']] 

recipe_id = recipes[~recipes['id'].isin(reviews['recipe_id'].unique())].iloc[0]['id'] 
row = merged[merged['id'] == recipe_id] 

print(merged) 
print(row)
```

Задание 5.2

```python
df = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='left')
grouped_df = df.groupby(['id', 'name']).size().reset_index(name='review_count')

recipe_id_no_reviews = recipes[~recipes['id'].isin(reviews['recipe_id'].unique())].iloc[0]['id']
row_for_recipe_no_reviews = grouped_df[grouped_df['id'] == recipe_id_no_reviews]

print(grouped_df)
print(row_for_recipe_no_reviews)
```

Задание 5.3

```python
df = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='left')
df['date_added_year'] = pd.to_datetime(df['submitted']).dt.year

mean_ratings = df.groupby(['date_added_year', 'name'])['rating'].mean().reset_index()
min_mean_rating = mean_ratings.groupby('date_added_year')['rating'].min()
worst_year = min_mean_rating.idxmin()
worst_recipes_in_worst_year = mean_ratings[mean_ratings['date_added_year'] == worst_year]

print(worst_year)
print(worst_recipes_in_worst_year)
```
